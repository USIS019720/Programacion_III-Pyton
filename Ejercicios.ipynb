```python
import pandas as pd
import matplotlib.pyplot as plt

archivo = pd.read_csv("/content/drive/MyDrive/tarea ejercicio progra III/tarjetas_por_sexo.csv", sep=",", encoding="utf-8")
print(archivo)

proceso =pd.crosstab(index=archivo["Gender"],columns="frecuencia")
#print(proceso)

frecuencia = archivo.loc[archivo.index>=0]["frecuencia"]
#print(frecuencia)
plt.pie(frecuencia, labels=frecuencia.index)
plt.xlabel("Utilizacion de tarjetas por sexo")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

archivo = pd.read_csv("/content/drive/MyDrive/tarea ejercicio progra III/total_dls_ciudad_tipo_tarjeta.csv", sep=",", encoding="utf-8")
print(archivo)

proceso =pd.crosstab(index=archivo["City"],columns="sum")
#print(proceso)

sum = archivo.loc[archivo.index]["sum"]
#print(Frecuencia)
plt.pie(sum, labels=sum.index)
plt.xlabel("Ciudad tipo de tarjeta")
plt.show()
```
